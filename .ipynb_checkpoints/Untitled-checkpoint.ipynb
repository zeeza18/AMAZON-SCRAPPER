```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
link = 'https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
page = requests.get(link)
```

```python
page
```

```python
page.content
```

```python
soup = bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('span', class_='a-profile-name')
names = list(dict.fromkeys(names))
```

```python
name = []
for i in range(0, len(names)):
    name.append(names[i].get_text())
```

```python
review_title= soup.find_all('a',class_='review-title-content')
```

```python
title = []
for i in range(0, len(review_title)):
    title.append(review_title[i].get_text().strip('\n'))
title
```

```python
rate = soup.find_all('i',class_='review-rating')
```

```python
rating = []
for i in range(2,len(rate)):
    rating.append(rate[i].get_text())
rating
```

```python
reviews = soup.find_all("span",{"data-hook":"review-body"})
```

```python
review= []
for i in range(0, len(reviews)):
    review.append(reviews[i].get_text().strip('\n'))
review
```

```python
data = pd.DataFrame()
```

```python
data['Customer_Name'] = name
data['Review_Title'] = title
data['Ratings'] = rating
data['Review'] = review
```

```python
data.head()
```

```python
data.to_csv('amzreview.csv', index=False)
```
